## Project Goal
For this assignment, I will use a labeled dataset of spam and non-spam emails to build a document classification model. The goal is to train the model on existing labeled documents and then evaluate how well it predicts whether new documents are spam or ham.

```python
import pandas as pd

# Load dataset (we will download it next)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data"

df = pd.read_csv(url, header=None)

print(df.shape)
df.head()
```

```python
# Check basic info about the dataset
df.info()
```

```python
# Separate features and target
X = df.iloc[:, :-1]   # all columns except last
y = df.iloc[:, -1]    # last column

print(X.shape)
print(y.shape)
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split

# Split the data (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

# Create the model
model = GaussianNB()

# Train the model
model.fit(X_train, y_train)

print("model trained")
```

```python
from sklearn.metrics import accuracy_score

# Make predictions
y_pred = model.predict(X_test)

# Check accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## Model Evaluation

The model achieved an accuracy of approximately 0.825, which indicates strong overall performance.

The confusion matrix shows that the model is effective at correctly identifying spam emails, with a high number of true positives. However, there are some false positives, where non-spam emails are incorrectly classified as spam.

This suggests that while the model is reliable for detecting spam, it may occasionally flag legitimate emails as spam. Overall, the model demonstrates a good balance but could be improved to reduce false positives.
